```python
import numpy as np
import pandas as pd
import cv2 as cv 
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt
from google.colab import files

#urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
 #      "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
 #     "https://placekitten.com/800/571","https://mysuru.acm.org/anand1.png"]  
# Read and display the image
# loop over the image URLs, you could store several image urls in the list

urls = ["https://mysuru.acm.org/anand1.png"]

for url in urls:
  image = io.imread(url) 
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  #final_frame = cv.hconcat((image, image_2))
  cv2_imshow(image_2)
  print('\n\n')
  invert=cv.bitwise_not(image_2)

  blur=cv.GaussianBlur(invert,(21,21),0)
  invertblur=cv.bitwise_not(blur)
  sketch=cv.divide(image_2,invertblur,scale=115)
  cv2_imshow(sketch)
  cv.imwrite("6.jpg",sketch)
  plt.imsave('5.jpg', sketch)
  #files.download("1.jpg")
  #plt.savefig("1.jpg")# empty outputs
 # files.download("1.jpg")
```
